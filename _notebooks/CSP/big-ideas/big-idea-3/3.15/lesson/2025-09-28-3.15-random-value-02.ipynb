# üóùÔ∏è Level 2 Lock ‚Äî Randomness & Logic

To break through, you must prove mastery over **RANDOM values** and **basic conditionals**.

‚öîÔ∏è **Your Task**: Solve each puzzle one by one. Only a correct answer advances you.

```python
%%html 
<div id="er-l2" style="margin:18px 0; padding:16px; border:1px solid rgba(255,255,255,.12); border-radius:14px; background:linear-gradient(180deg,#101635,#0b1026);">
  <h3 style="margin:0 0 10px;">üîí Level 2 Lock ‚Äî Exam-Style Practice</h3>
  <p style="margin:6px 0 12px; color:#cfe7ff">
    One at a time. Answer correctly to proceed.<br>
    ‚Ä¢ <em>Range</em>: ‚ÄúWhat values can <code>RANDOM(a,b)</code> produce?‚Äù (comma-separated integers)<br>
    ‚Ä¢ <em>MCQ</em>: Pick the correct percentage / condition.
  </p>

  <!-- Progress + Counter -->
  <div id="er-l2-progress" style="margin:10px 0 12px;">
    <div style="display:flex; align-items:center; gap:10px;">
      <div id="er-l2-counter" style="min-width:110px; font-weight:600;">Q 1 / 14</div>
      <div style="flex:1; height:8px; background:rgba(255,255,255,.12); border-radius:999px; overflow:hidden;">
        <div id="er-l2-bar" style="height:100%; width:0%; background:#7CFC00;"></div>
      </div>
    </div>
  </div>

  <!-- Puzzle area -->
  <div id="er-l2-puzzles"></div>

  <!-- Nav -->
  <div style="display:flex; gap:8px; margin-top:12px;">
    <button id="er-l2-prev" disabled>‚óÄ Back</button>
    <button id="er-l2-next" disabled>Next ‚ñ∂</button>
  </div>

  <!-- Status -->
  <div id="er-l2-status" style="margin-top:14px; font-weight:600;"></div>

  <style>
  /* General text */
  #er-l2, 
  #er-l2 label, 
  #er-l2 h4, 
  #er-l2 .puzzle, 
  #er-l2 .choices label, 
  #er-l2 code {
    color: #e9eefc; /* light off-white text */
  }

  /* Inputs */
  #er-l2 input[type="text"] {
    background: #0f1530;
    color: #e9eefc;
    border: 1px solid rgba(255,255,255,.25);
    padding: 4px 6px;
    border-radius: 6px;
  }

  /* Radio buttons with labels */
  #er-l2 input[type="radio"] {
    accent-color: #8ab4ff; /* modern browsers support accent color */
  }

  /* Buttons */
  #er-l2 button {
    padding:6px 10px;
    border-radius:6px;
    border:1px solid rgba(255,255,255,.25);
    background:rgba(255,255,255,.06);
    color:#e9eefc; /* force bright */
    cursor:pointer;
  }
  #er-l2 button:hover {
    background: rgba(255,255,255,.12);
  }
  #er-l2 button[disabled] {
    opacity:.5; 
    cursor:not-allowed;
  }

  /* Feedback colors */
  #er-l2 .feedback { margin-top:6px; font-size:.9rem; }
  #er-l2 .correct { color:#7CFC00; }
  #er-l2 .wrong { color:#FF6B6B; }

  /* Choice labels */
  #er-l2 .choices { display:grid; gap:4px; margin:6px 0; }
  #er-l2 .choices label {
    cursor: pointer;
  }

  /* Badge styling */
  #er-l2 .badge {
    display:inline-block; 
    font-size:.78rem; 
    padding:2px 6px; 
    border-radius:999px; 
    background:rgba(255,255,255,.1); 
    color:#cfe7ff;
    margin-left:6px;
  }
</style>


  <script>
    (function(){
      const puzzles = [
        // Your original range puzzles
        { id:"R1", type:"range", a:2,  b:5,  answer:[2,3,4,5],
          prompt: p => `Range ‚Ä¢ What outputs can <code>RANDOM(${p.a}, ${p.b})</code> give?` },
        { id:"R2", type:"range", a:0,  b:3,  answer:[0,1,2,3],
          prompt: p => `Range ‚Ä¢ What outputs can <code>RANDOM(${p.a}, ${p.b})</code> give?` },
        { id:"R3", type:"range", a:-2, b:2,  answer:[-2,-1,0,1,2],
          prompt: p => `Range ‚Ä¢ What outputs can <code>RANDOM(${p.a}, ${p.b})</code> give?` },

        // MCQs (Level 2)
        { id:"Q1", type:"mcq",
          stem:`What percent of runs display TRUE? <br><code>DISPLAY(RANDOM(5, 9) &gt; 9)</code>`,
          choices:["0%","20%","60%","100%"], answerIndex:0,
          explain:"RANDOM(5,9) yields only 5‚Äì9, so it‚Äôs never > 9 ‚Üí 0%."
        },
        { id:"Q2", type:"mcq",
          stem:`What percent of runs display TRUE? <br><code>DISPLAY(RANDOM(1, 10) ‚â§ 6)</code>`,
          choices:["9%","10%","60%","100%"], answerIndex:2,
          explain:"Ten equally likely outcomes (1‚Äì10); six of them ‚â§ 6 ‚Üí 6/10 = 60%."
        },
        { id:"Q3", type:"mcq",
          stem:`What percent of runs display TRUE? <br><code>DISPLAY(RANDOM(1, 5) = 5 OR RANDOM(1, 5) = 9)</code>`,
          choices:["20%","40%","60%","100%"], answerIndex:0,
          explain:"‚Äò=9‚Äô is impossible; ‚Äò=5‚Äô happens 1/5 ‚Üí 20%."
        },
        { id:"Q4", type:"mcq",
          stem:`What percent of runs display TRUE? <br><code>DISPLAY(RANDOM(1, 5) = 6 AND RANDOM(1, 5) ‚â§ 5)</code>`,
          choices:["0%","40%","60%","100%"], answerIndex:0,
          explain:"‚Äò=6‚Äô is impossible, so AND is always false ‚Üí 0%."
        },
        { id:"Q5", type:"mcq",
          stem:`What percent of runs display TRUE? <br><code>DISPLAY(RANDOM(1, 5) = 6 OR RANDOM(1, 5) ‚â§ 5)</code>`,
          choices:["0%","40%","60%","100%"], answerIndex:3,
          explain:"Left is impossible; right is always true ‚Üí OR is always true ‚Üí 100%."
        },
        { id:"Q6", type:"mcq",
          stem:`What percent of runs display TRUE? <br><code>DISPLAY(RANDOM(1, 10) = 6)</code>`,
          choices:["9%","10%","60%","100%"], answerIndex:1,
          explain:"One favorable out of ten equally likely ‚Üí 10%."
        },
        { id:"Q7", type:"mcq",
          stem:`What percent of runs display TRUE? <br><code>DISPLAY(RANDOM(5, 9) = 6)</code>`,
          choices:["10%","20%","60%","100%"], answerIndex:1,
          explain:"Five outcomes (5‚Äì9); one equals 6 ‚Üí 1/5 = 20%."
        },
        { id:"Q8", type:"mcq",
          stem:`What percent of runs display TRUE? <br><code>DISPLAY(RANDOM(5, 9) = 4)</code>`,
          choices:["0%","20%","60%","100%"], answerIndex:0,
          explain:"4 is outside 5‚Äì9 ‚Üí impossible ‚Üí 0%."
        },
        { id:"Q9", type:"mcq",
          stem:`What percent of runs display TRUE? <br><code>DISPLAY(RANDOM(5, 9) ‚â§ 9)</code>`,
          choices:["0%","20%","60%","100%"], answerIndex:3,
          explain:"All values 5‚Äì9 satisfy ‚â§ 9 ‚Üí 100%."
        },
        { id:"Q10", type:"mcq",
          stem:`A user inputs 5 for <code>a</code>. What is a possible value of <code>c</code>?<br>
                <code>a ‚Üê INPUT()</code><br>
                <code>b ‚Üê RANDOM(a, 6)</code><br>
                <code>c ‚Üê b ‚àí 6</code>`,
          choices:["‚àí3","‚àí1","1","3"], answerIndex:1,
          explain:"With a=5, b‚àà{5,6} ‚Üí c‚àà{‚àí1,0}. Only ‚àí1 appears."
        }
      ];

      const root = document.getElementById('er-l2');
      const mount = root.querySelector('#er-l2-puzzles');
      const status = root.querySelector('#er-l2-status');
      const prevBtn = root.querySelector('#er-l2-prev');
      const nextBtn = root.querySelector('#er-l2-next');
      const counter = root.querySelector('#er-l2-counter');
      const bar = root.querySelector('#er-l2-bar');

      const solved = new Set();     // ids solved correctly
      let idx = 0;                  // current puzzle index
      let lastAnsweredCorrect = false;

      function updateProgress(){
        counter.textContent = `Q ${idx+1} / ${puzzles.length}`;
        const pct = (solved.size / puzzles.length) * 100;
        bar.style.width = pct.toFixed(1) + '%';
        status.textContent = solved.size === puzzles.length
          ? "üîì All tumblers aligned ‚Äî the lock opens!"
          : `üîê ${solved.size}/${puzzles.length} solved.`;
        prevBtn.disabled = (idx === 0);
        nextBtn.disabled = !lastAnsweredCorrect || (idx === puzzles.length-1);
        nextBtn.textContent = (idx === puzzles.length-1) ? "Finish" : "Next ‚ñ∂";
      }

      function render(){
        mount.innerHTML = '';
        const p = puzzles[idx];
        const box = document.createElement('div');
        box.className = 'puzzle';
        if(p.type === 'range'){
          box.innerHTML = `
            <h4>Puzzle ${p.id} <span class="badge">Range</span></h4>
            <div>${p.prompt ? p.prompt(p) : `What outputs can <code>RANDOM(${p.a}, ${p.b})</code> give?`}</div>
            <input type="text" id="ans-${p.id}" placeholder="e.g. ${p.answer.join(',')}">
            <button id="check-${p.id}">Check</button>
            <div id="fb-${p.id}" class="feedback"></div>
          `;
          box.querySelector(`#check-${p.id}`).onclick = ()=>{
            const ans = box.querySelector(`#ans-${p.id}`).value
              .split(',').map(s=>s.trim()).filter(Boolean).map(Number);
            const corr = p.answer.slice().sort((a,b)=>a-b);
            const ok = ans.length===corr.length && corr.every(v=>ans.includes(v));
            const fb = box.querySelector(`#fb-${p.id}`);
            if(ok){
              fb.innerHTML = `‚úÖ <span class="correct">Correct.</span> Inclusive integers from <code>${p.a}</code> to <code>${p.b}</code>.`;
              fb.className = 'feedback correct';
              solved.add(p.id);
              lastAnsweredCorrect = true;
            } else {
              fb.innerHTML = `‚ùå <span class="wrong">Try again.</span> Hint: <code>RANDOM(a,b)</code> returns an <em>integer</em> in the inclusive range [a,b].`;
              fb.className = 'feedback wrong';
              lastAnsweredCorrect = false;
            }
            updateProgress();
          };
        } else {
          // MCQ
          const name = `mcq-${p.id}`;
          box.innerHTML = `
            <h4>Puzzle ${p.id} <span class="badge">MCQ</span></h4>
            <div>${p.stem}</div>
            <div class="choices">
              ${p.choices.map((c,i)=>`
                <label><input type="radio" name="${name}" value="${i}"> ${String.fromCharCode(65+i)}. ${c}</label>
              `).join('')}
            </div>
            <button id="check-${p.id}">Check</button>
            <div id="fb-${p.id}" class="feedback"></div>
          `;
          box.querySelector(`#check-${p.id}`).onclick = ()=>{
            const selected = box.querySelector(`input[name="${name}"]:checked`);
            const fb = box.querySelector(`#fb-${p.id}`);
            if(!selected){
              fb.textContent = "Pick an option.";
              fb.className = 'feedback wrong';
              lastAnsweredCorrect = false;
              updateProgress();
              return;
            }
            const ok = Number(selected.value) === p.answerIndex;
            if(ok){
              fb.innerHTML = `‚úÖ <span class="correct">Correct.</span> ${p.explain || ''}`;
              fb.className = 'feedback correct';
              solved.add(p.id);
              lastAnsweredCorrect = true;
            } else {
              fb.innerHTML = `‚ùå <span class="wrong">Not quite.</span>`;
              fb.className = 'feedback wrong';
              lastAnsweredCorrect = false;
            }
            updateProgress();
          };
        }
        mount.appendChild(box);
        // if already solved, unlock Next
        lastAnsweredCorrect = solved.has(p.id);
        updateProgress();
      }

      prevBtn.onclick = ()=>{
        if(idx > 0){ idx--; render(); }
      };
      nextBtn.onclick = ()=>{
        if(idx < puzzles.length - 1){ idx++; render(); }
        else {
          // Finished
          status.textContent = "üéâ Chamber cleared! You‚Äôve solved every tumbler.";
        }
      };

      render();
    })();
  </script>
</div>

```
